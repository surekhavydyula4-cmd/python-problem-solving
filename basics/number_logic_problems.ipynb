```python
# Leap Year
```

```python
def check_leap_year(yr):
  if (yr%4==0 and yr%100!=0) or yr%400==0:
    return "Leap year"
  return "Not leap year"
check_leap_year(2016)
```

```python
# Prime Number
```

```python
def check_prime(num):
  if num<=1:
    return "Not prime"
  else:
    for i in range(2,num):
      if num%i==0:
        return "Not prime"
    return "prime"
a=check_prime(23)
a
```

```python
# Palindrome Number
```

```python
def check_palindrome(num):
  temp=num
  pal_num=0
  while temp>0:
    r=temp%10
    temp=temp//10
    pal_num=(pal_num*10)+r
  if num==pal_num:
    return "Palindrome"
  return "Not palindrome"
check_palindrome(121)
```

```python
# Armstrong Number(the sum of its digits raised to the power of the total number of digits is equal to the number itself.)
```

```python
def check_arm(num):
  temp=num
  total=0
  n=len(str(num))
  while temp>0:
    r=temp%10
    temp=temp//10
    total+=r**n  # 1**3+5**3+3**3
  if total==num:
    return "Armstrong "
  return "Not armstrong"
check_arm(153)
```

```python
# Factorial
```

```python
def fact(n):
  temp=1
  if n<=1:
    return 1
  else:
    temp=n*fact(n-1)
  return temp
fact(5)
```

```python
# Sum of Digits
```

```python
def sum_digits(num):
  total=0
  while num>0:
    r=num%10
    num=num//10
    total+=r
  return total
a=sum_digits(123)
print(a)
```

```python
# Simple Calculator
```

```python
def cal(a,b,op):
  if op=='+':
    return a+b
  elif op=='-':
    return a-b
  elif op=='*':
    return a*b
  elif op=='/':
    return a/b
  else:
    return a%b
cal(12,34,'+')
```

```python
# Perfect number (sum of factors=num)
```

```python
def check_perfect(num):
  total=0
  for i in range(1,num):
    if num%i==0:
      total+=i
  if num==total:
    return 'Perfect Number'
  return 'Not a perfect Number'
check_perfect(6)
```

```python
# strong num(sum of factorial of digits =num)
```

```python
def check_strong(num):
  temp=num
  total=0
  while temp>0:
    fact=1
    r=temp%10
    temp=temp//10
    for i in range(1,r+1):
      fact*=i
    total+=fact
  if total==num:
    return 'Strong number'
  return 'Not a strong number'
check_strong(145)
```

```python
# Harshad Number(number divisible by sum of digits)
```

```python
def check_harshad_num(num):
  if num==0:
    return 'Not hasrhad num'
  temp=num
  total=0
  while temp>0:
    r=temp%10
    temp=temp//10
    total+=r
  if num%total==0:
    return 'Harshad number'
  return 'Not harshad number'
check_harshad_num(0)
```

```python
# Neon Number(sum of digits of square = number)
```

```python
def check_neon_num(num):
  temp=num*num
  total=0
  while temp>0:
    r=temp%10
    temp=temp//10
    total+=r

  if num==total:
    return 'Neon number'
  return 'Not neon number'
check_neon_num(9)
```

```python
# Spy Number(sum of digits = product of digits)
```

```python
def check_spy(num):
  total=0
  prd=1
  while num>0:
    r=num%10
    num=num//10
    total+=r
    prd*=r
  if total==prd:
    return "Spy num"
  return "Not a spy num"
check_spy(1124)
```

```python
# Duck Number(contains 0 but not at starting)
```

```python
def check_duck(num):
  temp=str(num)
  c=0
  if temp.startswith('0'):
    return "Not Duck"
  else:
    for i in temp:
      if i=='0':
        c+=1
    if c>=1:
      return "Duck"
    return "Not Duck"
check_duck(23)
```

```python
# Automorphic Number(square ends with same number)
```

```python
def chekc_morpic(num):
  sqr=num*num
  if str(sqr).endswith(str(num)):
    return "Automorph"
  return "Not Automorph"
chekc_morpic(25)
```

```python
def check_automorphic(num):
  sqr = num * num
  temp = num
  while temp > 0:
    if temp % 10 != sqr % 10:
      return "Not Automorph"
    temp //= 10
    sqr //= 10
  return "Automorph"

print(check_automorphic(25))
```
